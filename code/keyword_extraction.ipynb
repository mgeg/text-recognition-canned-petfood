```python
import os
import zipfile
```

```python
with zipfile.ZipFile("petfood.zip","r") as zip_ref:
    zip_ref.extractall("")
```

```python
filelist = []
for root, dirs, files in os.walk("/content/petfood"):
  for file in files:
    filelist.append("petfood/"+str(file))
```

```python
with open(filelist[0]) as f:
  doc = f.read()
```

```python
print(doc)
```

## KeyBERT

```python
from keybert import KeyBERT

kw_model = KeyBERT()
```

```python
kw_model.extract_keywords(doc, keyphrase_ngram_range=(1, 4),  use_mmr=True, diversity=0.7, top_n=5)
```

## Yake

```python
#!pip install git+https://github.com/LIAAD/yake
import yake

kw_extractor = yake.KeywordExtractor(dedupLim= 0.5,top= 5)
```

```python
kw_extractor.extract_keywords(doc)
```

## keyword extraction for all txt files

```python
for file in filelist:
  with open(file) as f:
    doc = f.read()
  
  keybert_words = kw_model.extract_keywords(doc, keyphrase_ngram_range=(1, 4), use_mmr=True, diversity=0.7, top_n=5)
  yake_words = kw_extractor.extract_keywords(doc)
  
  output_filename = file.split(".")[0]
  with open(str(output_filename) + '_output.txt', 'w') as f:
    f.write('Key Words:\n')
    f.write('\n'.join(keybert_words))
    f.write('\n'.join(yake_words))
```
